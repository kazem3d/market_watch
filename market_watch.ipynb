در این برنامه اطلاعات قسمت دیدبان بازار سایت بورس تهران گرفته شده و در یک فایل اکسل ذخیره می گردد.

```python
import pandas as pd
import requests
```

```python
url='http://www.tsetmc.com/tsev2/data/MarketWatchPlus.aspx'
```

```python
r=requests.get(url)
```

```python
main_text=r.text
csvs=main_text.split('@')
csv=csvs[2]
rows=csv.split(';')
row=[r.split(',') for r in rows]
```

```python
df=pd.DataFrame(row)
```

```python
# filter only stocks ticker
stock_keys=['300','303','305','309','380']
stock_filt=df[22].isin(stock_keys)
df=df.loc[stock_filt]
```

```python
# renaming columns name
col_rename={
    0:'tse_url',
    1:'en_ticker',
    2:'ticker',
    3:'name',
    5:'open',
    6:'price',
    7:'last',
    8:'tedad',
    9:'vol',
    10:'val',
    11:'day range_low',
    12:'day range_high',
    13:'yesterday',
    14:'eps',
    15:'base_vol',
    17:'bazar_number',
    18:'group_code',
    19:'max_range',
    20:'min_range',
    21:'tedade_kole_saham',
    22:'stock_code', 
}
df=df.rename(columns=col_rename)
```

```python
# group_code_list=df['group_code'].unique()
```

```python
# calculating p/e
df['price']=df['price'].astype(int)
# because some 'eps' are not number we use to_numeric and coerce
df['eps']=pd.to_numeric(df['eps'],errors='coerce')
# fill 'eps' that  are not number with zero
df['eps']=df['eps'].fillna(0)
df['eps']=df['eps'].astype(int)

df['p.e']=(df['price']/df['eps']).round(1)
```

```python
#calculate market cap 
df['tedade_kole_saham']=pd.to_numeric(df['tedade_kole_saham'])
df['market_cap']=(df['price']*df['tedade_kole_saham'])
```

```python
# replace persian letters with arabic
df['ticker']=df['ticker'].str.replace("ي","ی")
df['ticker']=df['ticker'].str.replace("ك","ک")
```
